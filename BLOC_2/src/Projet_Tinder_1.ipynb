```python
# Visualisation du DATAFRAME ages des personnes identifiées et ages de leurs partenaires
df_global_age=df[["age","age_o"]]
df_global_age.head()
```

```python
## visualisation par age des personnes identifiées

from matplotlib import pyplot as plt
df_global_age['age'].plot(kind='hist', bins=20, title='Age des personnes identifiées')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# visualisation par age des partenaires

from matplotlib import pyplot as plt
df_global_age['age_o'].plot(kind='hist', bins=20, title='Age des partenaires')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
#Etude des volumes Homme / Femmes
df["gender"] = df["gender"].replace({0: "F", 1: "M"})
dfcomptagegender = df["gender"].value_counts()

# Accéder aux comptes pour "M" et "F" de la série dfcomptagegender
M_nb = dfcomptagegender["M"] # Obtenir le nombre de "M" (hommes)
F_nb = dfcomptagegender["F"] # Obtenir le nombre de "F" (femmes)

print(f"Il y a {M_nb} Hommes")
print(f"Il y a {F_nb} Femmes")


print(dfcomptagegender)
```

```python
# dataframe age / ages partenaire / genre
df_age_gend = df[["age","age_o","gender"]]
df_age_gend.head()
```

```python
# visualisation de la recherche de partenaires en fonction du sexe et de l'age

fig_age_gend= px.box(df_age_gend, x="age", y="age_o", color="gender", color_discrete_map={"M": "blue", "F": "red"})
fig_age_gend.show()
```
