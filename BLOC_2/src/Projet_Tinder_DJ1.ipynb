## AXE 1: la différence d'ages est-elle un critère favorisant les matchs ?

```python
# visualisation de la recherche de partenaires en fonction du sexe et de l'age

fig_age_gend= px.box(df, x="age", y="age_o", color="gender", color_discrete_map={"Hommes": "blue", "Femmes": "salmon"})
fig_age_gend.show()
```

On se rends compte que les femmes ont partenaires du même age legerement plus agées. Pour hommes les partenaires sont souvent plus jeunes, parfois nettement.


```python
# Création d'une varible différence d'ages
df['age_diff'] = df['age_o'] - df['age']

# Conversion binaire de la variable "match"
df['is_match'] = df['match'].map({'Oui': 1, 'Non': 0})

# Regroupement par différence d'âge
age_match = df.groupby('age_diff')['is_match'].mean().reset_index()

# Graphique
plt.figure(figsize=(10, 6))
sns.lineplot(data=age_match, x='age_diff', y='is_match')
plt.axvline(0, color='gray', linestyle='--', label='Même âge')
plt.title('Taux de match selon la différence d’âge')
plt.xlabel('Différence d’âge (partenaire - soi)')
plt.ylabel('Probabilité de match')
plt.legend()
plt.grid(True)
plt.show()
```

Les matches sont les plus fréquents quand la différence d’âge entre partenaires est proche de zéro (±2 ans).
Le taux de match reste élevé pour de petites différences d’âge dans les deux sens, mais chute fortement au-delà de ±10 ans.
Des pics extrêmes apparaissent aux extrémités, mais sont probablement dus à un nombre d’observations trop faible pour être fiables.

Par concequent la différence d'age entre partenaires n'est pas une variable considérable pour qu'il est des Matchs

## AXE 2 : L'ordre des rencontres est-elle une variable importante pour qu'il y ait match ?

```python
# 1. Calcul des pourcentages de match par ordre
match_counts = df.groupby(['order', 'match']).size().unstack(fill_value=0)
match_counts['total'] = match_counts.sum(axis=1)
match_counts['match_rate'] = match_counts.get('Oui', 0) / match_counts['total'] * 100

# 2. Créer la figure avec un axe secondaire
fig = make_subplots(specs=[[{"secondary_y": True}]])

# 3. Histogrammes pour "Oui" et "Non"
for match_value, color in {"Oui": "blue", "Non": "salmon"}.items():
    subset = df[df["match"] == match_value]
    fig.add_trace(
        go.Histogram(
            x=subset["order"],
            name=match_value,
            marker_color=color,
            opacity=0.6,
            nbinsx=20
        ),
        secondary_y=False
    )

# 4. Ajouter la ligne de pourcentage de match
fig.add_trace(
    go.Scatter(
        x=match_counts.index,
        y=match_counts['match_rate'],
        name="Taux de match (%)",
        mode='lines+markers',
        line=dict(color='green', width=2)
    ),
    secondary_y=True
)

# 5. Mise en page
fig.update_layout(
    title_text="Importance de l'ordre des rencontres pour qu'il y ait match",
    barmode="overlay",
    xaxis_title="Ordre de la rencontre",
    yaxis_title="Nombre de personnes",
    yaxis2_title="Taux de match (%)",
    title_x=0.5,
    legend=dict(x=0.8, y=1.1)
)

fig.show()




```

1 Le taux de match est le plus élevé lors des premiers rendez-vous (ordre faible): Les utilisateurs semblent plus enclins à matcher au début de leur parcours.
2 On observe une baisse progressive du taux de match avec l’augmentation du nombre de rencontres : Cela pourrait refléter une fatigue, une exigence accrue ou une baisse d’intérêt.
3 Le volume de rencontres reste élevé, mais leur efficacité diminue.
4 Les pics ou creux aux extrémités doivent être interprétés avec prudence (peu de données).

Conclusion : Apparement, l’ordre chronologiquedes rencontre joue un rôle dans la probabilité de match.
